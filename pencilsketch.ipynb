```python
#importing libraries
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Loading Original Image
img = cv2.imread("pup.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(13,13))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
#Converting Image to GrayScale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(13,13))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
#Inverting the Image
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(13,13))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
#Blurring/Smoothing the Image
img_blur = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(12,12))
plt.imshow(img_blur,cmap="gray")
plt.axis("off")
plt.title("Blurred Image")
plt.show()
```

```python
#Converting your images to pencil sketches
final = cv2.divide(img_gray, 255 - img_blur, scale=255)
plt.figure(figsize=(13,13))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
# viewing all Images in sequence
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_blur,cmap="gray")
plt.axis("off")
plt.title("blurred Image")
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
